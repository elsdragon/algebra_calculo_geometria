## 1. Multiconjuntos

Este ejercicio pondr치 a prueba tu habilidad resolver un problema usando vectores.

**Problema**: Implementar las operaciones de los multiconjuntos (utilizando las librer칤as y estructuras de datos vistas en el curso).

**Datos:**

Un multiconjunto es un conjunto en el que un elemento puede repetirse, es decir, cada elemento posee una multiplicidad (un n칰mero natural) que indica cu치ntas veces el elemento es miembro del conjunto. Por ejemplo, en el multiconjunto `{a, a, b, b, b, c}`, las multiplicidades de los miembros a, b, y c son 2, 3, y 1, respectivamente.

Al igual que los conjuntos, poseen las siguientes caracter칤sticas y operaciones:
- Cardinalidad: indica el n칰mero de elementos del multiconjunto. Por ejemplo, la cardinalidad del multiconjunto `{a, a, b, b, b, c}` es 6 (la suma de sus multiplicidades).
- Inserci칩n: permite insertar una ocurrencia de un elemento en el multiconjunto.
- Eliminaci칩n: permite eliminar una ocurrencia de un elemento del multiconjunto.
- Comparaci칩n: compara dos multiconjuntos para determinar si son iguales.
- Pertenencia: determina si un elemento pertenece al multiconjunto.
- Subconjunto: determina si un multiconjunto es subconjunto de otro.
- Uni칩n: conjunci칩n de todos los elementos de dos multiconjuntos (sumando sus multiplicidades si un elementos est치 en los dos).
- Intersecci칩n: elementos que est치n en los dos multiconjuntos qued치ndonos con la multiplicidad m치s peque침a.
- Diferencia: restar a un multiconjunto los elementos de otro.

```python
### Crear una funci칩n que dada una lista devuelva un multiconjunto
### El multiconjunto que devuelve puede crearse con la estructura de datos que se quiera (incluso una lista)

### DEVOLVEMOS UN DICCIONARIO (CLAVE: los elementos  y VALOR: el n칰mero de ocurrencias)

def multiconjunto(elementos: list)-> dict:
    """
    Devuelve un multiconjunto con la lista de elementos dada
    
    Argumentos:
        elementos -- lista de elementos
        
    Ejemplo:
        elementos = [1,1,1,3,3,1,4,5,1,5]
    """
    
    #Definir la variable de retorno como diccionario vacio
    multiconjunto = {}
    for element in elementos:
        # Si el elemento ya est치 como clave sumar uno al valor
        if element in multiconjunto:
            multiconjunto[element]+=1
        else:
            # Si el valor no est치 como clave a침adirlo e inicializar el valor a 1
            multiconjunto[element] = 1
                
    # Devolver el dictionario resultante
    return multiconjunto
    
    
```

```python
### Crear una funci칩n que dado un multiconjunto devuelva su cardinalidad

def cardinalidad(multiconjunto: dict)-> int:
    """
    Devuelve la cardinalidad del multiconjunto dado
    
    Argumentos:
        multiconjunto -- multiconjunto devuelto por la funci칩n creada anteriormente
    """
    # Sumar los valores de todas las claves del diccionario
    return sum(multiconjunto.values())
```

```python
### Crear una funci칩n que dado un multiconjunto y un elemento devuelva el multiconjunto con el elemento insertado

def inserta(multiconjunto: dict, elemento: object)-> dict:
    """
    Devuelve el multiconjunto habiendo insertado el elemento dado
    
    Argumentos:
        multiconjunto -- multiconjunto devuelto por la funci칩n creada anteriormente
        elemento -- elemento a insertar
    """
    # Comprobamos si el elemento a insertar ya existe como clave en el multiconjunto
    if elemento in multiconjunto:
        # Si existe sumamos 1 al valor de la clave
        multiconjunto[elemento] += 1
    else:
        # Si no existe a침adimos la clave y asignamos 1 al valor
        multiconjunto[elemento] = 1
        
    # Devolvemos el multiconjunto resultante
    return multiconjunto
```

```python
### Crear una funci칩n que dado un multiconjunto y un elemento devuelva el multiconjunto con el elemento eliminado

def elimina(multiconjunto: dict, elemento: object)-> dict:
    """
    Devuelve el multiconjunto habiendo eliminado una ocurrencia del elemento dado
    
    Argumentos:
        multiconjunto -- multiconjunto devuelto por la funci칩n creada anteriormente
        elemento -- elemento a eliminar
    """
    # Comprobamos si el elemento es una las claves del multiconjunto
    if elemento in multiconjunto:
        # Si es una de las claves que reste al valor 1
        multiconjunto[elemento] -= 1
        # Comprobamos si el valor de esa clave es 0
        if multiconjunto[elemento] == 0:
            # Si es cero eleminamos la clave
            del multiconjunto[elemento]
    # Devolvemos el multiconjunto modificado       
    return multiconjunto
```

```python
### Crear una funci칩n que dado un multiconjunto y un elemento devuelva si el elemento pertenece al multiconjunto

def pertenece(multiconjunto: dict, elemento: object)-> bool:
    """
    Devuelve si el elemento pertenece al multiconjunto
    
    Argumentos:
        multiconjunto -- multiconjunto devuelto por la funci칩n creada anteriormente
        elemento -- elemento a determinar si pertenece
    """
    return elemento in multiconjunto
```

```python
### Crear una funci칩n que dados dos multiconjuntos devuelva si el primero es subconjunto del segundo

def subconjunto(multiconjunto1: dict, multiconjunto2:dict)-> bool:
    """
    Devuelve si multiconjunto1 es subconjunto de multiconjunto2
    
    Argumentos:
        multiconjunto1 -- multiconjunto devuelto por la funci칩n creada anteriormente
        multiconjunto2 -- multiconjunto devuelto por la funci칩n creada anteriormente
    """
    # Incializamos la variable de resultado a True
    result = True
    # Iteramos por las claves del multiconjunto1
    for clave in multiconjunto1:
        # Comprobamos que cada clave del multiconjunto1 cumple 2 condiciones:
        # 1. Que la clave est치 en el multiconjunto2
        # 2. Que el valor de la clave en el multiconjunto1 es menor o igual al valor de esa clave en el multiconjunto2
        result = pertenece(multiconjunto2, clave) and multiconjunto1[clave] <= multiconjunto2[clave]
        # Si el resultado es falso salimos del bucle
        if result == False:
            return result
    # Devolvemos el resultado
    return result
```

```python
### Crear una funci칩n que dados dos multiconjuntos devuelva si son iguales

def iguales(multiconjunto1: dict, multiconjunto2: dict)-> bool:
    """
    Devuelve si multiconjunto1 es igual a multiconjunto2
    
    Argumentos:
        multiconjunto1 -- multiconjunto devuelto por la funci칩n creada anteriormente
        multiconjunto2 -- multiconjunto devuelto por la funci칩n creada anteriormente
    """
    # Inicializamos una variable resultado a False
    result = False
    # Iteramos por las claves del multiconjunto1
    for clave in multiconjunto1:
        # Comprobamos 2 condiciones:
        # 1. Que la clave del multiconjunto1 est칠 en el multiconjunto2
        # 2. Que el valor de la clave en el multiconjunto1 sea igual al valor de la clave en el multiconjunto2
        result = pertenece(multiconjunto2, clave) and multiconjunto1[clave]==multiconjunto2[clave]
        # Si el resultado es falso salimos del bucle
        if result == False:
            return result
    # Devolvemos el resultado
    return result
```

```python
### Crear una funci칩n que dados dos multiconjuntos devuelva su uni칩n


def union(multiconjunto1: dict, multiconjunto2: dict)-> dict:
    """
    Devuelve la uni칩n de los multiconjuntos
    
    Argumentos:
        multiconjunto1 -- multiconjunto devuelto por la funci칩n creada anteriormente
        multiconjunto2 -- multiconjunto devuelto por la funci칩n creada anteriormente
    """
    # Inicializamos una variable como un diccionario vacio
    newmulti = {}
    # Iteramos por la clave del multiconjunto1
    for clave1 in multiconjunto1:
        # Comprobamos si est치 en el multiconjunto2
        if pertenece(multiconjunto2, clave1):
            # Si est치 guardamos la clave sumando los valores de esa clave en los dos multiconjuntos
            newmulti[clave1] = multiconjunto1[clave1]+multiconjunto2[clave1]
        else:
            # Si no est치 guardamos la clave con el valor del multiconjunto1
            newmulti[clave1] = multiconjunto1[clave1]
    # Iteramos sobre las claves del multiconjunto2
    for clave2 in multiconjunto2:
        # Comprobamos si la clave no est치 en el 1(si est치 en el 1 ya le hemos tratado en el caso de arriba)
        if  not pertenece(multiconjunto1, clave2):
            # Si no est치 en el multiconjunto1 guardamos en nuestro nuevo diccionario la clave con su valor
            newmulti[clave2] = multiconjunto2[clave2]
    # Devolvemos como multiconjunto el diccionario resultante
    return newmulti
```

```python
### Crear una funci칩n que dados dos multiconjuntos devuelva su intersecci칩n

def interseccion(multiconjunto1: dict, multiconjunto2: dict)-> dict:
    """
    Devuelve la intersecci칩n de los multiconjuntos
    
    Argumentos:
        multiconjunto1 -- multiconjunto devuelto por la funci칩n creada anteriormente
        multiconjunto2 -- multiconjunto devuelto por la funci칩n creada anteriormente
    """
    # Inicializamos una variable como un diccionario vacio
    newmulti = {}
    # Iteramos por la clave del multiconjunto1
    for clave in multiconjunto1:
        # Comprobamos si est치 en el multiconjunto2 para a침adirlo al conjunto
        # Si pertenece al multiconjunto 2 comprobamos tambi칠n los valores
        if pertenece(multiconjunto2, clave) and  multiconjunto1[clave]<=multiconjunto2[clave]:
            newmulti[clave]= multiconjunto1[clave]
            
        elif pertenece(multiconjunto2, clave) and  multiconjunto1[clave]> multiconjunto2[clave]:
            newmulti[clave]= multiconjunto2[clave]
                
    # Devolvemos el diccionario resultante
    return newmulti
```

```python
### Crear una funci칩n que dados dos multiconjuntos devuelva su diferencia

def diferencia(multiconjunto1: dict, multiconjunto2: dict)-> dict:
    """
    Devuelve la diferencia de los multiconjuntos
    
    Argumentos:
        multiconjunto1 -- multiconjunto devuelto por la funci칩n creada anteriormente
        multiconjunto2 -- multiconjunto devuelto por la funci칩n creada anteriormente
    """
    # Inicializamos una variable como un diccionario vacio
    newmulti = {}
    # Iteramos por el multiconjunto1 por su clave y valor
    for clave, valor in multiconjunto1.items():
        # Comprobamos si la clave no est치 en el multiconjunto2
        if not pertenece(multiconjunto2, clave):
            # Si no est치 guardamos en la variable la clave valor
            newmulti[clave]=valor
        # En el caso de que si est칠 esa clave en el multiconjunto2
        # Comprobamos que la resta de valores sea mayor que 0
        elif multiconjunto1[clave]-multiconjunto2[clave]>0:
            # Si cumple la condici칩n guardamos la clave y como valor la resta
            newmulti[clave]=multiconjunto1[clave]-multiconjunto2[clave]
    # Devolvemos el diccionario resultante                
    return newmulti
```

Deberas crear tu propia funci칩n para calcular el error de reconstrucci칩n, que viene definido por:

$$SSE =  \sum_{n}^{i=1}  \begin{Vmatrix}x_{i} -  \widehat{x}_i \end{Vmatrix} ^2 $$

Donde:

- $x_i$ son los valores de la matriz original X
- $\widehat{x}_i$ son los valores de la matriz reconstruida

```python
### Funci칩n para calcular el error de reconstrucci칩n

import numpy as np

def sse_score(X, X_hat):
    """
    Funci칩n para calcular el error de reconstrucci칩n
    
    Argumentos:
        X -- Matriz Original
        X_hat -- Matriz Reconstruida
        
    Ejemplo:
        X = np.array([[1, 2], [3, 4]])
        X_hat = np.array([[1.01, 1.75], [2.81, 3.99]])
        sse = sse_score(X, X_hat) # -> 0.09879
    """
    return np.sum((X-X_hat)**2)

X = np.array([[1, 2], [3, 4]])
X_hat = np.array([[1.01, 1.75], [2.81, 3.99]])
sse = sse_score(X, X_hat) # -> 0.09879
sse
```

Una vez que ya tenemos la funci칩n `sse` hecha, podemos pasar a construir la funci칩n que ejecutar치 `SVM`.

```python
### Funci칩n para ejecutar SVM
### Tiene como entrada una matriz X
### Devuelve U, s, Vt

### Hint: S debe ser una matriz diagonal


def svm(X):
    """
    Funci칩n que ejecuta SVM y devuelve U, S, Vt
    
    Argumentos:
        X -- Matriz Original
        
    Ejemplo:
        X = np.array([[1, 2], [3, 4]])
        U, S, Vt = svm(X)  
        
        # -> U = array([[-0.40455358, -0.9145143 ],
        #               [-0.9145143 ,  0.40455358]])
        # -> S = array([[5.4649857 , 0.        ],
        #               [0.        , 0.36596619]])   
        # -> Vt = array([[-0.57604844, -0.81741556],
        #                [ 0.81741556, -0.57604844]])          
    """ 
    U, S, Vt = np.linalg.svd(X, full_matrices=False)
    S = np.diag(S)
    return U, S, Vt

X = np.array([[1, 2], [3, 4]])
U, S, Vt = svm(X)  
```

Como hemos visto en clase, las matrices obtenidas a partir de `SVM` nos sirven para reconstruir la matriz original `X`. Para ello, construye una funci칩n que permita reconstruir la matriz original `X` a partir de `U, s, Vt`.

```python
### Funci칩n para reconstruir la matriz original a partir de U, s, Vt
### Tiene como entrada U, s, Vt
### Devuelve X_hat

def reconstruction(U, S, Vt):
    """
    Funci칩n que reconstruye la matriz original a partir de U, s, Vt
    
    Argumentos:
        U -- Matriz de Singular Vectors
        s -- Matriz de Eigenvalues
        Vt -- Matriz de Singular Vectors
        
    Ejemplo:
        U = np.array([[-0.40455358, -0.9145143 ],
                      [-0.9145143 ,  0.40455358]])
        S = np.array([[5.4649857 , 0.        ],
                      [0.        , 0.36596619]])
        Vt = np.array([[-0.57604844, -0.81741556],
                       [ 0.81741556, -0.57604844]])
        X_hat = reconstruction(U, S, Vt)
        
        # X_hat -> array([[0.99999999, 1.99999998],
        #                 [3.00000003, 4.00000001]])
    """
    return U@S@Vt
```

```python
U = np.array([[-0.40455358, -0.9145143 ],
              [-0.9145143 ,  0.40455358]])
S = np.array([[5.4649857 , 0.        ],
              [0.        , 0.36596619]])
Vt = np.array([[-0.57604844, -0.81741556],
               [ 0.81741556, -0.57604844]])
X_hat = reconstruction(U, S, Vt)
X_hat

# X_hat -> array([[0.99999999, 1.99999998],
#                 [3.00000003, 4.00000001]])
```

Calcula el error de reconstrucci칩n usando la funci칩n `sse` que has programado anteriormente.

```python
sse = sse_score(X, X_hat)
sse
```

```python
### Funci칩n que recibe una imagen A y devuelve la imagen comprimida
### Tiene como entrada A y el n칰mero de componentes para realizar la reducci칩n de dimensionalidad
### Devuelve la imagen comprimidad, el error de reconstrucci칩n y el ratio de compresi칩n

### Hint: Usa las funciones anteriormente construidas



def image_compression(A, n_components):
    """
    Funci칩n para comprimir una imagen A
    
    Argumentos:
        A -- Imagen original
        n_components -- N칰mero de componentes
            
    Ejemplo:
        A_hat, sse, comp_ratio = image_compression(A, n_components=50)
    """
    U, S, Vt = svm(A)
    U_reducida = U[:, :n_components]
    S_reducida = S[:n_components, :n_components]
    Vt_reducida = Vt[:n_components, :]
    
    A_hat = reconstruction(U_reducida, S_reducida, Vt_reducida) 
       
    comp_ratio =(A.shape[0]*n_components + n_components +n_components *A.shape[1]) / (A.shape[0]*A.shape[1])
    
    
    sse = sse_score(X, X_hat)
    return A_hat, sse, comp_ratio

```

#### Subset del dataframe por valores

El siguiente paso es construir una funci칩n que recibe un `data_frame`, el nombre de una columna, un valor m칤nimo y un valor m치ximo `cutoffs`. Nos devuelve un dataframe excluyendo las filas donde el valor de la columna indica est치 fuera de los valores m칤nimos y m치ximos que le hemos indicado.

```python
### Funci칩n para crear un nuevo subset en base a valores
### Como argumento recibe un dataframe y una lista de tuples
### Tuples: (column_name, min_value, max_value)
### Devuelve un dataframe que excluye las filas donde los valores, en la columna que le hemos indicado, exceden los valores
### que le hemos indicado
### No eliminar la fila si los valores son iguales al min/max valor

def column_cutoff(data_frame, cutoffs):
    """Crea un nuevo dataframe en base a unos l칤mites
    
    Argumentos:
        data_frame -- Dataframe Object
        cutoffs -- Lista de tuples con el siguiente formato:
        (column_name, min_value, max_value)
        
    Ejemplo:
        data_frame = read_into_data_frame('train.csv')
        # Remove data points with SalePrice < $50,000
        # Remove data points with GrLiveAre > 4,000 square feet
        cutoffs = [('SalePrice', 50000, 1e10), ('GrLivArea', 0, 4000)]
        selected_data = column_cutoff(data_frame, cutoffs)
    """
    for i in range(len(cutoffs)):
        column_value = cutoffs[i][0]
        min_value = cutoffs[i][1]
        max_value = cutoffs[i][2]
        data_frame = data_frame[(data_frame[column_value]>=min_value)&(data_frame[column_value]<=max_value)]
                
    return data_frame
```

#### M칤nimos Cuadrados / Least Squares

Ahora, implementar치s la ecuaci칩n $w_{LS}$:

<center>$w_{LS} = (X^T X)^{1}X^T y,$</center>

```python
### Funci칩n para resolver la ecuaci칩n wLS
### Toma como argumentos dos matrices, una para X y otra para y
### Asumimos que las matrices tienen las dimensiones correctas

### Paso 1: Asegurate que n > d. 
### Es decir, que el n칰mero de observaciones es mayor que el n칰mero de dimensiones.
### O lo que es lo mismo, que el n칰mero de filas de cada matriz sea mayor que el n칰mero de columnas
### Si no es as칤, debes transponer las matrices

### Paso 2: Debes a침adir a la matriz X un vector columna del tama침o (n x 1)

### Paso 3: Usa la ecuaci칩n de arriba para obtener wLS



def least_squares_weights(input_x, target_y):
    """Resuelve la ecuaci칩n para wLS
    
    Argumentos:
        input_x -- Matriz con los datos de entrenamiento
        target_y -- Vector con los datos de salida
        
    Ejemplo:
        import numpy as np
        training_y = np.array([[208500, 181500, 223500, 
                                140000, 250000, 143000, 
                                307000, 200000, 129900, 
                                118000]])
        training_x = np.array([[1710, 1262, 1786, 
                                1717, 2198, 1362, 
                                1694, 2090, 1774, 
                                1077], 
                               [2003, 1976, 2001, 
                                1915, 2000, 1993, 
                                2004, 1973, 1931, 
                                1939]])
        weights = least_squares_weights(training_x, training_y)
        
        print(weights)  #--> np.array([[-2.29223802e+06],
                        #              [ 5.92536529e+01],
                        #              [ 1.20780450e+03]])
                           
        print(weights[1][0]) #--> 59.25365290008861
    
    Asumimos:
        -- target_y es un vector con el mismo n칰mero de observaciones que input_x
    """
    observaciones = input_x.shape[0]
    dimensiones = input_x.shape[1]
    
    if not observaciones>dimensiones:
        input_x = input_x.T
        target_y = target_y.T
        
    vector_columna = np.ones((input_x.shape[0],1))
    
    X = np.concatenate((vector_columna, input_x), axis = 1)
    
    
    return (np.linalg.inv(X.T@X))@X.T@target_y
    
        
    
    
```

```python
training_y = np.array([[208500, 181500, 223500, 
                        140000, 250000, 143000, 
                        307000, 200000, 129900, 
                        118000]])
training_x = np.array([[1710, 1262, 1786, 
                        1717, 2198, 1362, 
                        1694, 2090, 1774, 
                        1077], 
                       [2003, 1976, 2001, 
                        1915, 2000, 1993, 
                        2004, 1973, 1931, 
                        1939]])
weights = least_squares_weights(training_x, training_y)

print(weights)  #--> np.array([[-2.29223802e+06],
                #              [ 5.92536529e+01],
                #              [ 1.20780450e+03]])

print(weights[1][0]) #--> 59.25365290008861
```

## 4. Linear Regression - Gradient Descent

En este ejercicio resolveras el mismo problema anterior pero usando **Gradient Descent**

**Objetivos**:
- Asegurar los fundamentos matem치ticos detr치s del Gradient Descent.

**Problema**: Usando datos sobre el precio de la vivienda, intentaremos predecir el precio de una casa en base a la superficie habitable con un modelo de regresi칩n.

**Datos:** [Kaggle's House Prices Dataset](https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data)

**Repaso:**

$$ RSS(w) = \sum_{n=1}^{N}[y_n-f(x_n)]^2 =  \sum_{n=1}^{N}[y_n- (w_0 + \sum_{d=1}^{D}w_dx_{nd}) ]^2 .$$

Loss function:

$$ RSS(w) = \frac{1}{2}\sum_{n=1}^{N}[y_n-f(x_n)]^2$$

Y lo que queremos es minimizar esta distancia, para que el modelo se acerque lo m치ximo posible a los valores verdaderos.

$$\nabla RSS(w) = X^T(Xw^t-y)$$

En resumen, el gradient descendiente para una regresi칩n lineal, se basa en resolver esta ecuaci칩n de forma iterativa:

$$w^{t+1} = w^t - \eta * \nabla RSS(w)$$

#### Leer Datos

```python
import pandas as pd
import numpy as np

# Leer datos
data = pd.read_csv('./data/train.csv')

# Extraer dichas columnas
newData = data[['GrLivArea','SalePrice']]
print(newData.head())

# Contruir x - y
x = newData['GrLivArea']
y = newData['SalePrice']

# Standarizar los datos
x = (x - x.mean()) / x.std()
x = np.c_[np.ones(x.shape[0]), x] 

print("Shape of X: ", x.shape)
print("Shape of y:", y.shape)
```

#### Gradient Descent

```python
### Funci칩n para encontrar los valores w usando Gradient Descent
### Toma como argumentos: X, y, w, n_iterations, eta
### Completa la funci칩n a침adiendo la loss funci칩n y la updating rule


def gradient_descent(x, y, w, iterations, eta):
    """Gradient descent
    
    Argumentos:
        x -- Matriz con los datos de entrenamiento
        y -- Vector con los datos de salida
        w -- Vector aleatoriamente inicializado
        iterations -- N칰mero de iteraciones
        eta -- Learning Rate
        
    Ejemplo:
        import numpy as np

        # Learning rate
        eta = 0.01 

        # N칰mero de iteraciones
        iterations = 2000 #No. of iterations

        # Seed para inicializar w
        np.random.seed(123)
        w0 = np.random.rand(2)
        
        training_y = np.array([208500, 181500, 223500, 
                                140000, 250000])
        training_x = np.array([[ 1.        ,  0.37020659],
                               [ 1.        , -0.48234664],
                               [ 1.        ,  0.51483616],
                               [ 1.        ,  0.38352774],
                               [ 1.        ,  1.29888065]])
                            
        weights, loss = gradient_descent(training_x, training_y, w0, iterations, eta)
        
        print(weights[-1])  #--> np.array([183845.82320222  40415.66453324])
    """
    weights = []
    loss = []
    
    for i in range(iterations+1):
        
        RSS = x.T@(x@w-y)
        
        w = w - eta* RSS
        
        weights.append(np.array(w))
        
        loss_actual = np.sum((y - x@w)**2) / 2
               
        loss.append(loss_actual)
    
    return weights, loss
    
```

```python
# Learning rate
eta = 0.01 

# N칰mero de iteraciones
iterations = 2000 #No. of iterations

# Seed para inicializar w
np.random.seed(123)
w0 = np.random.rand(2)

training_y = np.array([208500, 181500, 223500, 
                        140000, 250000])
training_x = np.array([[ 1.        ,  0.37020659],
                       [ 1.        , -0.48234664],
                       [ 1.        ,  0.51483616],
                       [ 1.        ,  0.38352774],
                       [ 1.        ,  1.29888065]])

weights, loss = gradient_descent(training_x, training_y, w0, iterations, eta)

print(weights[-1])  #--> np.array([183845.82320222  40415.66453324])
type(weights[-1])
```

```python
import re
import lxml
from bs4 import BeautifulSoup
import urllib
import urllib.request

url = "https://es.wikipedia.org/wiki/Canis_lupus_familiaris"

def parse_from_url(url):
    """
    Funci칩n para extraer el contenido (raw text) de una p치gina web
    """
    
    html = urllib.request.urlopen(url).read()
    soup = BeautifulSoup(html, "html.parser" )
    for script in soup(["script", "style"]):
        script.extract()
        
    text = soup.get_text()
    
    # Eliminar saltos de linea
    text = re.sub('\s+', ' ', text)
    return text

parse_from_url(url)[:1000]
```

- Los perros son maravillosos...
- Los maravillosos a침os 80...

Por tanto, el conjunto para estas dos frases ser칤a `{"los", "perros", "son", "maravillosos", "a침os", "80"}`. Debemos realizar esto para todas las combinaciones posibles, es decir:

- web_1

En este caso, las combinaciones ser칤an (no importa el orden) `[web_1, web_2]`, `[web_1, web_3]`, `[web_2, web_3]`

```python
### Escribe una funci칩n que recibe una lista de N p치ginas web
### Y calcula todas las combinaciones posibles entre ellas, no importa el orden
### [web_1, web_2, web_3, ...]
### Devuelve una lista de tuples con las combinaciones [(web_1, web_2), (web_1, web_3), ...]

# HINT: Puedes implementar esta funci칩n como quieras pero la librer칤a itertools 
#       proporciona una funci칩n llamada `combinations` para realizar esta tarea.


import itertools
    
def combinations(url_ls):
    """Calcula todas las combinaciones posibles entre los elementos de una lista
    
    Argumentos:
        url_ls -- Lista de urls
        
    Ejemplo:
        url_ls = ['https://es.wikipedia.org/wiki/Canis_lupus_familiaris', 
        'https://es.wikipedia.org/wiki/Canis_lupus',
        'https://es.wikipedia.org/wiki/Felis_silvestris_catus']
        
        permutation = combinations(url_ls)  
    
    Requerimientos:
        Puedes implementar esta funci칩n como quieras pero la librer칤a itertools 
        proporciona una funci칩n llamada `combinations` para realizar esta tarea.
    """
    
    return list(itertools.combinations(url_ls, 2))

url_ls = ['https://es.wikipedia.org/wiki/Canis_lupus_familiaris', 
        'https://es.wikipedia.org/wiki/Canis_lupus',
        'https://es.wikipedia.org/wiki/Felis_silvestris_catus']
        
permutation = combinations(url_ls)  
print(permutation)
```

```python
### Escribe una funci칩n que recibe una lista con tuples
### [({'que', 'hola'}, {'que', 'es', 'guay'}), ({'que', 'hola'}, {'madrid', 'la', 'es'})]
### Y devuelve una lista con la union de los conjuntos
### [({'que', 'hola', 'es', 'guay'}), ({'que', 'hola', 'madrid', 'la', 'es'})]

def union(comb_ls):
    """Calcula la uni칩n por cada tuple de una lista 
    
    Argumentos:
        comb_ls -- Lista de tuples
        
    Ejemplo:
        comb_ls = [({'que', 'hola'}, {'que', 'es', 'guay'}), ({'que', 'hola'}, {'madrid', 'la', 'es'})]
        
        union_ls = union(comb_ls)  # -> [{'es', 'que', 'hola', 'guay'}, {'es', 'que', 'hola', 'madrid', 'la'}]
    """
    lista = []
    for tupla in comb_ls:
        conjuntotemp = set()
        for conjunto in tupla:
            conjuntotemp.update(conjunto)
        lista.append(conjuntotemp)
        
    return lista

comb_ls = [({'que', 'hola'}, {'que', 'es', 'guay'}), ({'que', 'hola'}, {'madrid', 'la', 'es'})]
        
union_ls = union(comb_ls) 
print(union_ls) # -> [{'es', 'que', 'hola', 'guay'}, {'es', 'que', 'hola', 'madrid', 'la'}]
```

Una vez que tenemos una lista de conjuntos por cada par de p치ginas web, podemos convertir el texto de la p치gina web a un vector.

```python
def set2vector(tokens_web1, tokens_web_2, set_web1, set_web2):
    """
    Funci칩n para convertir un conjunto a vector
    
    Argumentos:
        tokens_web1 -- Contenido tokenizado de p치gina web 1
        tokens_web_2 -- Contenido tokenizado de p치gina web 2
        set_web1 -- Conjunto de palabras de la p치gina web 1
        set_web2 -- Conjunto de palabras de la p치gina web 2
        
    Ejemplo:
        tokens_web1 = ["hola", "que", "tal", "soy", "Alex"]
        tokens_web_2 = ["hola", "me", "llamo"]
        set_web1 = {"hola", "que", "tal"}
        set_web2 = {"hola", "me", "llamo"}
        union_ls = set2vector(tokens_web1, tokens_web_2, set_web1, set_web2)  
        
    Requerimientos:
        Depende de la funci칩n `union()`
    """
    
    # Unimos los conjuntos
    join_set = union([(set_web1, set_web2)])[0]
    
    web1_array = []
    web2_array = [] 

    for word in join_set:
        if word in tokens_web1:
            web1_array.append(1)
        else:
            web1_array.append(0)
        if word in tokens_web_2:
            web2_array.append(1)
        else:
            web2_array.append(0)

    return np.array(web1_array), np.array(web2_array)

tokens_web1 = ["hola", "que", "tal", "soy", "Alex"]
tokens_web_2 = ["hola", "me", "llamo"]
set_web1 = {"hola", "que", "tal", "soy", "Alex"}
set_web2 = {"hola", "me", "llamo"}
web1_array, web2_array = set2vector(tokens_web1, tokens_web_2, set_web1, set_web2)  
```

### Cosine Similarity

Por 칰ltimo, ya podemos implementar la ecuaci칩n: $\cos \theta = \frac{\overrightarrow{u} \cdot \overrightarrow{v}}{|\overrightarrow{u}||\overrightarrow{v}|}$

```python
### Escribe una funci칩n que recibe dos vectores, u y v
### Y devuelva la similaridad entre ambos vectores
###
### Paso 1: Si u y v son listas -> Convertirlo a arrays
###
### Paso 2: Calcula la similaridad entre ambos vectores


import numpy as np

def cosine_similarity(u, v):
    """Calcula la similaridad entre dos vectores
    
    Argumentos:
        u -- Vector 1
        v -- Vector 2
        
    Ejemplo:
        u = np.array([1, 2, 3])
        v = np.array([3, 2, 1])
        
        similarity = cosine_similarity(u, v)  # -> 0.71428
    """
    return (u@v)/(np.linalg.norm(u)* np.linalg.norm(v))
    

u = np.array([1, 2, 3])
v = np.array([3, 2, 1])
        
similarity = cosine_similarity(u, v)  # -> 0.71428
similarity
    
```
